```python
# Lets test out my theory. if n and n + 1 have the same (or extremly close) time of sale, then I can consider them to be the same order where n is rows with item = Flavor and n + 1 are rows with item = the syrup the flavor belongs to
sales_clean = sales_clean.reset_index(drop=True)
flavorRows = sales_clean[sales_clean["Item"].str.contains("flavor", case=False)].index # create a list of indexes with rows of flavor aka n
#print(flavorRows) #sanity check, this has 588 rows
flavorSyrupRows = flavorRows + 1 # this is indexes of hypthetical drinks without flavor
#print(flavorSyrupRows) #  check



from datetime import datetime, date

paired_with = {}  # i -> the index of its paired drink row

for i in flavorRows:
    has_next = (i + 1) in sales_clean.index
    has_prev = (i - 1) in sales_clean.index and i - 1 >= 0

    diff_next = None
    diff_prev = None

    if has_next:
        t_i = datetime.combine(date.today(), sales_clean.loc[i, 'Time'])
        t_next = datetime.combine(date.today(), sales_clean.loc[i + 1, 'Time'])
        diff_next = abs((t_next - t_i).total_seconds())

    if has_prev:
        t_prev = datetime.combine(date.today(), sales_clean.loc[i - 1, 'Time'])
        t_i = datetime.combine(date.today(), sales_clean.loc[i, 'Time'])
        diff_prev = abs((t_i - t_prev).total_seconds())

    # decide which neighbor (if any) is the actual pair
    candidates = []
    if diff_next is not None and diff_next < 2:
        candidates.append((diff_next, i + 1))
    if diff_prev is not None and diff_prev < 2:
        candidates.append((diff_prev, i - 1))

    if candidates:
        candidates.sort()               # smallest gap first
        paired_with[i] = candidates[0][1]  # take the closer one

print(len(paired_with))

#print(len(matchesNext) + len(matchesPrev)) # 587 THATS ALMOST ALL OF THEM

# trying to find the other one
matched_indices = set()
for i in flavorRows:
    has_next = (i + 1) in sales_clean.index
    has_prev = (i - 1) in sales_clean.index and i - 1 >= 0
    matched = False
    if has_next:
        t_i = datetime.combine(date.today(), sales_clean.loc[i, 'Time'])
        t_next = datetime.combine(date.today(), sales_clean.loc[i + 1, 'Time'])
        if abs((t_next - t_i).total_seconds()) < 2:
            matched = True
    if has_prev:
        t_prev = datetime.combine(date.today(), sales_clean.loc[i - 1, 'Time'])
        t_i = datetime.combine(date.today(), sales_clean.loc[i, 'Time'])
        if abs((t_i - t_prev).total_seconds()) < 2:
            matched = True
    if matched:
        matched_indices.add(i)

unmatched = flavorRows.difference(matched_indices)
sales_clean.loc[unmatched]

sales_clean.loc[unmatched+1]  # should be either 9:07:08 if i-1 OR 9:09:28 if i + 1: result its i - 1 70 second difference thats a huge gap tho so i think its better to remove this row

sales_clean.drop(index=unmatched, inplace=True) # drop bad row

# Okay now that my theory has been proven, lets combine the drinks with their syrups

flavor = ""
for key, value in paired_with.items(): # This changes all those drinks to have the flavorings and updates its sales info as well
    flavor = sales_clean.loc[key, "Item"][:-7]
    gSales = sales_clean.loc[key, "Gross Sales"] + sales_clean.loc[value, "Gross Sales"]
    discount = sales_clean.loc[key, "Discounts"] + sales_clean.loc[value, "Discounts"] 
    nSales = sales_clean.loc[key, "Net Sales"] + sales_clean.loc[value, "Net Sales"]

    sales_clean.loc[value, "Modifiers Applied"] = sales_clean.loc[value, "Modifiers Applied"].fillna('') + ", " + flavor
    sales_clean.loc[value, "Gross Sales"] = gSales
    sales_clean.loc[value, "Discounts"] = discount
    sales_clean.loc[value, "Net Sales"] = nSales

    #print(sales_clean.iloc[32]) # this is an error so everything is good!!
sales_clean.drop(index = paired_with.keys(), inplace = True)
```
